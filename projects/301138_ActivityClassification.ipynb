Dataset information
===

>The experiments have been carried out with a group of 30 volunteers within an age bracket of 19-48 years. Each person performed six activities (WALKING, WALKING_UPSTAIRS, WALKING_DOWNSTAIRS, SITTING, STANDING, LAYING) wearing a smartphone (Samsung Galaxy S II) on the waist. Using its embedded accelerometer and gyroscope, we captured 3-axial linear acceleration and 3-axial angular velocity at a constant rate of 50Hz. The experiments have been video-recorded to label the data manually. The obtained dataset has been randomly partitioned into two sets, where 70% of the volunteers was selected for generating the training data and 30% the test data. 

> **Notes:**
- Features are normalized and bounded within [-1,1].
- Each [561] feature vector is a row on the text file.
- The units used for the accelerations (total and body) are 'g's (gravity of earth -> 9.80665 m/seg2).
- The gyroscope units are rad/seg.
- A video of the experiment including an example of the 6 recorded activities with one of the participants can be seen in the following link: http://www.youtube.com/watch?v=XOEN9W05_4A

Feature description
===
>The features selected for this database come from the accelerometer and gyroscope 3-axial raw signals tAcc-XYZ and tGyro-XYZ. These time domain signals (prefix 't' to denote time) were captured at a constant rate of 50 Hz. Then they were filtered using a median filter and a 3rd order low pass Butterworth filter with a corner frequency of 20 Hz to remove noise. Similarly, the acceleration signal was then separated into body and gravity acceleration signals (tBodyAcc-XYZ and tGravityAcc-XYZ) using another low pass Butterworth filter with a corner frequency of 0.3 Hz. 

>Subsequently, the body linear acceleration and angular velocity were derived in time to obtain Jerk signals (tBodyAccJerk-XYZ and tBodyGyroJerk-XYZ). Also the magnitude of these three-dimensional signals were calculated using the Euclidean norm (tBodyAccMag, tGravityAccMag, tBodyAccJerkMag, tBodyGyroMag, tBodyGyroJerkMag). 

>Finally a Fast Fourier Transform (FFT) was applied to some of these signals producing fBodyAcc-XYZ, fBodyAccJerk-XYZ, fBodyGyro-XYZ, fBodyAccJerkMag, fBodyGyroMag, fBodyGyroJerkMag. (Note the 'f' to indicate frequency domain signals). 

>These signals were used to estimate variables of the feature vector for each pattern:  \
_('-XYZ' is used to denote 3-axial signals in the X, Y and Z directions.)_
- tBodyAcc-XYZ \
- tGravityAcc-XYZ \
- tBodyAccJerk-XYZ \
- tBodyGyro-XYZ \
- tBodyGyroJerk-XYZ \
- tBodyAccMag \
- tGravityAccMag \
- tBodyAccJerkMag \
- tBodyGyroMag \
- tBodyGyroJerkMag \
- fBodyAcc-XYZ \
- fBodyAccJerk-XYZ \
- fBodyGyro-XYZ \
- fBodyAccMag \
- fBodyAccJerkMag \
- fBodyGyroMag \
- fBodyGyroJerkMag \

> The set of variables that were estimated from these signals are: \
- mean(): Mean value \
- std(): Standard deviation \
- mad(): Median absolute deviation \
- max(): Largest value in array \
- min(): Smallest value in array \
- sma(): Signal magnitude area \
- energy(): Energy measure. Sum of the squares divided by the number of values. \
- iqr(): Interquartile range \
- entropy(): Signal entropy \
- arCoeff(): Autorregresion coefficients with Burg order equal to 4 \
- correlation(): correlation coefficient between two signals \
- maxInds(): index of the frequency component with largest magnitude \
- meanFreq(): Weighted average of the frequency components to obtain a mean frequency \
- skewness(): skewness of the frequency domain signal \
- kurtosis(): kurtosis of the frequency domain signal  \
- bandsEnergy(): Energy of a frequency interval within the 64 bins of the FFT of each window. \
- angle(): Angle between to vectors. \

>Additional vectors obtained by averaging the signals in a signal window sample. These are used on the angle() variable: \
- gravityMean \
- tBodyAccMean \
- tBodyAccJerkMean \
- tBodyGyroMean \
- tBodyGyroJerkMean \

Methods of feature selection & extraction
===
- max_corr: \
  Choose features with highest absolute value of correlation with label, using Pearson's correlation of a sample.

- max_importance: \
  Choose features with highest importance based on a single unpruned decision tree. Here importance means frequency of use as a splitting criterium in tree nodes.

- PCA: \
  Substitue features for components computed by a PCA algorithm.

- kBest: \
  Choose features using sklearn method sklearn.feature_selection.SelectKBest with score function of mutual information. 

- all: \
  Control group with all 561 features intact. Optional, long computation time (especially with AdaBoost).


Models considered
===
- DT: \
  Single **decision tree** with optimized max depth D.

- RF: \
  **Random forest** with optimized number of trees N and max depth D

- KNN: \
  **K Nearest Neighbours** with optimized value of K

- rbf_SVC: \
  **Radial-basis function kernel Support Vector Classifier** with optimized value of C, one-vs-rest (since singular SVC estimator is binary). One-vs-rest means that for each class an estimator is trained than distinguishes between that class and all others.

- linearSVC: \
  **Linear Support Vector Classifier** with optimized value of C, one-vs-rest (since single SVC estimator is binary). 
- AdaBoost: \
  **AdaBoost with decision trees** of max depth 1 as base estimators and optimized number of estimators N and learning rate R.

###Methods of feature selection/extraction

```
# features most correlated to the label according to Pearson correlation (linear)
def max_corr(trainX, trainY, valX, testX, F):
    train_set = train_X.join(trainY, lsuffix='_caller', rsuffix='_other')

    corrmat = train_set.corr()
    corr_ranking = corrmat.iloc[:-1,-1:] # only in comparison with target label
    corr_ranking = corr_ranking.iloc[corr_ranking['activity'].abs().argsort()]
    features_max_corr = corr_ranking.iloc[-F:,:].index
    vals = np.absolute(corr_ranking.iloc[-F:,:].values.ravel())

    trainX = trainX.filter(features_max_corr)
    valX = valX.filter(features_max_corr)
    testX = testX.filter(features_max_corr)

    return features_max_corr, vals, trainX, valX, testX


# features with highest importance in a singular decision tree
def max_importance(trainX, trainY, valX, testX, F, md=9):
    model = sklearn.tree.DecisionTreeClassifier(max_depth=md)
    model.fit(trainX.values, trainY.values.ravel())

    feature_importances = pd.Series(model.feature_importances_, index=trainX.columns)
    features_max_imp = feature_importances.nlargest(F).index
    vals = feature_importances.nlargest(F).values.ravel()

    trainX = trainX.filter(features_max_imp)
    valX = valX.filter(features_max_imp)
    testX = testX.filter(features_max_imp)

    return features_max_imp, vals, trainX, valX, testX


# features extracted by PCA
def PCA(trainX, valX, testX, F):
    pca = sklearn.decomposition.PCA(n_components = F)
    pca.fit(trainX.values)

    trainX = pd.DataFrame(pca.transform(trainX.values))
    valX = pd.DataFrame(pca.transform(valX.values))
    testX = pd.DataFrame(pca.transform(testX.values))

    return ["p" + str(f + 1) for f in range(F)], pca.explained_variance_ratio_, trainX, valX, testX


# features selected based on their ANOVA F-value
def kBest(trainX, trainY, valX, testX, F):
    selector = sklearn.feature_selection.SelectKBest(k=F, score_func=sklearn.feature_selection.mutual_info_classif)
    selector.fit(trainX.values, trainY.values.ravel())

    trainX = trainX.iloc[:, selector.get_support(indices=True)]
    valX = valX.iloc[:, selector.get_support(indices=True)]
    testX = testX.iloc[:, selector.get_support(indices=True)]
    vals = sklearn.feature_selection.mutual_info_classif(trainX.values, trainY.values.ravel(), discrete_features=False, random_state=0)
    
    return trainX.columns, vals, trainX, valX, testX



all_features = lambda trainX, trainY, valX, testX, F: (trainX.columns, np.ones(trainX.shape[1]), trainX, valX, testX)
PCA_l = lambda trainX, trainY, valX, testX, F: PCA(trainX, valX, testX, F)

feature_choices = {
    'max corr' : max_corr, 
    'kBest' : kBest,
    'max importance' : max_importance, 
    'PCA' : PCA_l,
    'all' : all_features
}
```

###Classification models

Some helper functions for training models:

```
def train_model(trainX, trainY, model):
    model.fit(trainX.values, trainY.values.ravel())
    e_train = 1.0 - model.score(trainX.values, trainY.values.ravel())

    return model, e_train

def chooseParams(trainX, trainY, valX, valY, pnames, prange, model_type):
    chosen_model = None
    min_e_val = 1.1

    for p in prange:
        m, e_train = train_model(trainX, trainY, model_type(p))
        e_val = 1.0 - m.score(valX.values, valY.values.ravel())

        if e_val < min_e_val:
            min_e_val = e_val
            chosen_model = (pnames, p, m, e_train, e_val)

    return chosen_model
```

```
# singular decision tree
def decisionTree(trainX, trainY, valX, valY, minD=6, maxD=12):
    model_type = lambda md: sklearn.tree.DecisionTreeClassifier(max_depth=md, random_state=0)
    prange = list(range(minD, maxD + 1)) + [None]

    model = chooseParams(trainX, trainY, valX, valY, "D", prange, model_type)
    
    return model


# random forest
def randomForest(trainX, trainY, valX, valY, minD=6, maxD=12, minN=60, maxN=160):
    model_type = lambda ps: sklearn.ensemble.RandomForestClassifier(n_estimators=ps[0], max_depth=ps[1], random_state=0)
    prange = [(n, d) for n in range(minN, maxN + 1, 20) for d in range(minD, maxD + 1)]

    model = chooseParams(trainX, trainY, valX, valY, ("N", "D"), prange, model_type)
    
    return model


# K nearest neighbours
def KNN(trainX, trainY, valX, valY, minK=6, maxK=12):
    model_type = lambda k: sklearn.neighbors.KNeighborsClassifier(n_neighbors=k)
    prange = list(range(minK, maxK + 1))

    model = chooseParams(trainX, trainY, valX, valY, "K", prange, model_type)
    
    return model


# RBF kernel support vector classifier - one vs rest
def rbfSVC(trainX, trainY, valX, valY, minC=-6, maxC=1):
    model_type = lambda c: sklearn.svm.SVC(C=c, kernel='rbf', decision_function_shape='ovr', random_state=0)
    prange = [10**c for c in range(minC, maxC + 1)]

    model = chooseParams(trainX, trainY, valX, valY, "C", prange, model_type)
    
    return model


# linear support vector classifier - one vs rest
def linearSVC(trainX, trainY, valX, valY, minC=-6, maxC=1):
    model_type = lambda c: sklearn.svm.LinearSVC(C=c, multi_class='ovr', max_iter=3000, random_state=0)
    prange = [10**c for c in range(minC, maxC + 1)]

    model = chooseParams(trainX, trainY, valX, valY, "C", prange, model_type)
    
    return model

# adaboost with decision trees (default)
def AdaBoost(trainX, trainY, valX, valY, minN=50, maxN=200, minR=4, maxR=10):
    model_type = lambda ps: sklearn.ensemble.AdaBoostClassifier(n_estimators=ps[0], learning_rate=ps[1], random_state=0)
    prange = [(n, r / 10.0) for n in range(minN, maxN + 1, 25) for r in range(minR, maxR + 1)]

    model = chooseParams(trainX, trainY, valX, valY, ("N", "R"), prange, model_type)
    
    return model


models = {
    'DT' : decisionTree,
    'RF' : randomForest,
    'KNN' : KNN,
    'rbf_SVC' : rbfSVC,
    'linearSVC' : linearSVC,
    'AdaBoost' : AdaBoost
}
```

###Testing functions

```
def trainAndTest(chosen_model, trainX, trainY, valX, valY, testX, testY):
    pnames, params, model, e_train, e_val = chosen_model(trainX, trainY, valX, valY) 
    e_test = 1.0 - model.score(testX.values, testY.values.ravel())

    # calculate confusion matrix
    cm = sklearn.metrics.confusion_matrix(testY.values.ravel(), model.predict(testX.values))

    return (pnames, params, e_train, e_val, e_test, cm)


def runTests(F, trainX, trainY, valX, valY, testX, testY, verbose=False, skip_all=True):
    results = []
    cumulative_cm = None

    for fc_name, res in fc_datasets.items():
        if skip_all and fc_name == 'all':
            continue

        _, _, trX, vX, tsX = res
        
        for m_name, model in models.items():
            if verbose:
                print(fc_name, ", ", m_name, sep="")

            pnames, params, e_train, e_val, e_test, cm = trainAndTest(model, trX, trainY, vX, valY, tsX, testY)
            results += [(fc_name, m_name, params, e_train, e_val, e_test, cm)]

            if cumulative_cm is None:
                cumulative_cm = np.array(cm)
            else:
                cumulative_cm += np.array(cm)

            if verbose:
                print("chosen params =", pnames, "->", params)
                print("e_train =", e_train)
                print("e_val =", e_val)
                print("e_test =", e_test)
                print("confusion matrix (rows=actual, cols=pred):")
                print(cm)
                print()
        
        if verbose:
            print()

    return results, cumulative_cm
```

###Summary of results
**Methods** 

Based on error rate best possible feature selection method is max importance for a decision tree with error rate of 12-15% for tests. It is closely followed by PCA that has higher error rate for decision trees (single, random forest and adaboost), error rate 10-22% for tests. Worst of all is max correlation with label with average test error of about 36%.

Poor performance of max correlation and kBest methods suggest that many of the features of the dataset are highly correlated. What's more if such a subset of highly inter-correlated features is also highly correlated with label, then this methods may result in many redundant features being selected. This leads to loss of information and low accuracy of models.

---
**Models**

Error rate tells us that best model regardless of method is rbf kernel SVC, with average error of 16.4% for tests. Linear SVC is close second with ~17% average test error. By far the worst turned out to be AdaBoost with average error of 32%.

AdaBoost's performance might be affected by the high overlapping of classes and since this algorithm focuses on improving on previous missclassifications it probably led it to try and fit problematic regions (sitting vs standing, walking vs walking upstairs) therefore decreasing overalll accuracy of the model.

---
**Best combination**

Best possible combination of method and model (excluding control group containing all features) for this dataset is max importance method with rbf kernel SVC model achieving final error rate of 10.34% for tests (89.66% accuracy).
